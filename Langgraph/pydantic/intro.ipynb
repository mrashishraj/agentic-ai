```python
from pydantic import BaseModel
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="ashish", age=30, city="dataninja")
print(person)
```

```python
from dataclasses import dataclass

@dataclass
class test1:
    name: str
    age: int
    city: str
person = test1(name="Alice", age=30, city=30)
print(person)
```
