## For the daily push Use below Commond
- git add .
- git commit -m "your message"
- git push


## To start any new repo push 
## First install the git in the folder
- git init
- git add .
- git commit -m "add commit"
- git remote add origin https://github.com/{your-username}/{your-repo-name}
- git push -u origin main

## To check all .git folder
- find . -name ".git" -type d

## To find all file by it it's size in git commit
- find . -size +5M -not -path "./.git/*"
