```python
from pydantic import BaseModel
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="ashish", age=30, city="dataninja")
print(person)
```

```python
from dataclasses import dataclass

@dataclass
class test1:
    name: str
    age: int
    city: str
person = test1(name="Alice", age=30, city=30)
print(person)
```

#### 2. Model with optinal fields

Add optional fields using python's optional type:

Definition:
- Optional[type]: Indicates the field can be none
- Default value (= None or = True): Makes the field optional
- Required fields must still be provided
- pydantic validates types even for optional fields when values are provided






```python
from typing import Optional
class employee(BaseModel):
    id: int
    name: str
    age: int
    city: Optional[str] = None
    salary: Optional[float] = None

emp1 = employee(id=1, name="ashish", age=25, city="dataninja", salary=50000.0)
print(emp1)
```

```python
from pydantic import BaseModel

class Classroom(BaseModel):
    room_number: str
    students: list[str]
    capacity: int

classroom=Classroom(room_number="101", students=["ashish", "raj", "mitul",], capacity=30)
print(classroom)
```

#### 4. Model with nested models create complex structures with nested models:

Create complex structures with nested models:

```python
from pydantic import BaseModel

class address(BaseModel):
    street:str
    city:str
    zip_code:int

class customer(BaseModel):
    name:str
    age:int
    address:address

customerData = customer(
    name="ashish", 
    age=30, 
    address={"street": "123 Main St", "city": "kanpur", "zip_code": "12345"}
    )

print(customerData)
```

### Pydantic fields: customization and contraints

##### The field function in pydantic enhances model fields beyond basic type hints by allowing you to specify validation rules, default vlaues, aliases, and more. Here's a comprehensive tutorial with examples.


```python
from pydantic import BaseModel, Field

class test(BaseModel):
    name:str = Field(min_length=3, max_length=50)
    price:float = Field(gt=0,le=1000)
    quantity:int = Field(ge=1, le=100)

# Valid instance
item = test(name="Book", price=1, quantity=10)
print(item)
```

```python
from pydantic import BaseModel, Field

class user(BaseModel):
    username: str = Field(..., description="The unique username of the user")
    email: str = Field(default_factory=lambda:"user@example.com", description="The email address of the user")
    age: int = Field(default=18, description="The age of the user", ge=18, le=100)

User = user(username="ashish")
print(User)

User2 = user(username="raj")
print(User2)
```
