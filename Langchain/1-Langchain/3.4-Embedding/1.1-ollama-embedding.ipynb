# Ollama

#### Ollama supports embadding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompt with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
)

result = embeddings.embed_documents([
    "What is the second letter of greek alphabet "
])



print(result)
print(len(result[0]))  # Embedding dimension
```
