#### CREATE A TODO LIST TO KEEP TRACK OF TASKS

```python
to_do_list=["Buy groceries","Clean the house","Pay bills","Call mom"]

## Adding to task
to_do_list.append("Go for a walk")
to_do_list.append("Read a book")

##removing a task that is done
to_do_list.remove("Buy groceries")

##checking if a task is in the list
if "Clean the house" in to_do_list:
    print("Task found!")

print("To Do List remaining")
for task in to_do_list:
    print("- " + task)
```
